```python
import networkx as nx
import pickle
import pandas as pd
```

```python
from google.colab import drive
import os

drive.mount('/content/drive')

shared_folder = '/content/drive/MyDrive/Ariya Narayanasamy/Data/Pickle Saves'
os.chdir(shared_folder)
```

```python
with open('tm_graph.pickle', 'rb') as handle:
    G = pickle.load(handle)
```

```python
betweenness = nx.betweenness_centrality(G)
```

```python
data = [{
    'station_name': node, 'betweenness_centrality': centrality, 'lines': list(G.nodes()[node]['serviced_lines'].keys())
} for node, centrality in betweenness.items()]
df = pd.DataFrame(data)
```

```python
df = df.sort_values(by='betweenness_centrality', ascending=False)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('centrality_data.csv', index=False)
```
